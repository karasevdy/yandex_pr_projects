### Напишите функцию для вычисления итоговой sMAPE

#### Функция локального (не-итогового) sMAPE - станет scorer

```python
# на вход функция простого (не-итогового) sMAPE должна брать (target, predictions)
# ее буду использовать в качестве scorera в GridSearchCV и cross_val_score 

def smape(target, predictions):
    return 1/len(target) * np.sum(2*np.abs(target-predictions)/(np.abs(target)+np.abs(predictions))*100)
```

```python
# сделаем сразу из функции smape scorer, чтобы потом о нем не забывать

from sklearn.metrics import make_scorer
smape_scorer = make_scorer(smape, greater_is_better=False)
```

#### Функция итоговой sMAPE

```python
# на вход берет target_test_rough, predictions_test_rough, target_test_final, predictions_test_final



def smape_total(target_test_rough, predictions_test_rough, target_test_final, predictions_test_final):
    smape_rough = 1/len(target_test_rough) * np.sum(2*np.abs(target_test_rough-predictions_test_rough)/(np.abs(target_test_rough)+np.abs(predictions_test_rough))*100)
    smape_final = 1/len(target_test_final) * np.sum(2*np.abs(target_test_final-predictions_test_final)/(np.abs(target_test_final)+np.abs(predictions_test_final))*100) 
    return 0.25*smape_rough + 0.75*smape_final
    
```
